```python
plants = ['Spinach', 'Turnip','Rhubarb','Broccoli','Kale']
died = [10,25,5,30,21]
germinated = [74,88,56,69,59]
```

```python
plt.figure(figsize=(12,6))
plt.bar(plants, germinated, color='green',label='Germinated Seedlings')
plt.bar(plants, died, color='red', bottom=germinated, label=('Dead seedlings'))
plt.title('Germination Rates', c='black')
plt.ylabel('Number of seedlings')
plt.legend(fancybox=True, facecolor='black')
```

```python
plt.figure(figsize=(12,6))
plt.barh(plants, germinated, color='green',label='Germinated Seedlings')
plt.barh(plants, died, color='red',label=('Dead seedlings'))
plt.title('Germination Rates', c='black')
plt.ylabel('Number of seedlings')
plt.legend(fancybox=True, facecolor='black')
```

```python
titanic = pd.read_csv('../data/titanic.csv')
```

```python
titanic['age'] = pd.to_numeric(titanic['age'], errors='coerce')
```

```python
plt.hist(titanic['age'])
```

```python
first_class = titanic[titanic['pclass'] == 1]['age']
second_class = titanic[titanic['pclass'] == 2]['age']
third_class = titanic[titanic['pclass'] == 3]['age']
```

```python
plt.hist(first_class, label='first class', alpha=0.5)
plt.hist(second_class, label='second class', alpha=0.5)
plt.hist(third_class, label='third class', alpha=0.5)
plt.legend(labelcolor='black')
```

```python
nums = np.arange(5)
```

```python
plt.figure(figsize=(12,6), dpi=200, facecolor='black')
plt.suptitle('Xs')
plt.subplot(1,3,1)
plt.plot(nums)
plt.title('X')
plt.subplot(1,3,2)
plt.title('X**2')
plt.plot(nums**2, c='blue')
plt.subplot(1,3,3)
plt.title('X**3')
plt.plot(nums**3, c='yellow')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,10))
ax = plt.subplot(1,3,1)
plt.hist(first_class, label='first class', color='yellow')
plt.subplot(1,3,2, sharey=ax)
plt.hist(second_class, label='second class', color='red')
plt.subplot(1,3,3, sharey=ax)
plt.hist(third_class, label='third class', color='blue')

plt.legend(labelcolor='black')
plt.show()
```
