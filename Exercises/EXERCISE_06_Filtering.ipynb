# Filtering Data Exercise
Answer the following questions using the `bestsellers.csv` dataset

```python
import pandas as pd
books = pd.read_csv("data/bestsellers.csv")
```

```python
books.info()
```

```python
#Find the books written by Pete Souza
pete_souza = books["Author"] == "Pete Souza"
books[pete_souza]
```

```python
# Find the books that are under 10 dollars
under_10_dollars = books["Price"] < 10
books[under_10_dollars]
```

```python
#Find the books that have a price between 50 and 60 dollars
between_50_60_dollars = books["Price"].between(50,60)
books[between_50_60_dollars]
```

```python
#Find all the books written by Kristin Hannah, Andy Weir, or Delia Owens
authors = ["Kristin Hannah", "Andy Weir", "Delia Owens"]
books[books["Author"].isin(authors)]
```

```python
books.head()
```

```python
#Find the Non Fiction books that are rated 4.9
nonfiction = books["Genre"] == "Non Fiction"
four_point_nine_rating = books["User Rating"] == 4.9
books[nonfiction & four_point_nine_rating]
```

```python
#Find the fiction book with the lowest User Rating
fiction = ~nonfiction
books[fiction].sort_values("User Rating").head(1)
```

```python
#Find 2012's top 5 Fiction books with the most Reviews
year2012 = books["Year"] == 2012
books[year2012 & fiction].sort_values("Reviews", ascending=False).head(5)
```

```python
#Create a bar plot showing the 5 authors who have the most books with a rating under 4.5
rating_under4_point_5 = books["User Rating"] < 4.5
books[rating_under4_point_5]["Author"].value_counts().head(5).plot(kind="bar")
```
